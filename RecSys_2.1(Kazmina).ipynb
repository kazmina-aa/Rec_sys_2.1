```python
import pandas as pd
import numpy as np
import math
import json
```

```python
#вариант у меня 10, но строка 9
my_user = 10
v = my_user - 1
k = 4
```

```python
data = pd.read_csv('data.csv', delimiter = ', ', engine = 'python')
context_day = pd.read_csv('context_day.csv', delimiter = ', ', engine = 'python')
context_place = pd.read_csv('context_place.csv', delimiter = ', ', engine = 'python')
movie_names = pd.read_csv('movie_names.csv', delimiter = ', ', engine = 'python')
```

```python
data
```

```python
data = data.drop(data.columns[0],axis = 1)#удалили столбец с пользователями
data
```

```python
similarity_metric = np.zeros(data.shape[0])#создали нулевую матрицу для вычисления метрики сходства
similarity_metric
```

```python
without_evaluating = -1
for u in range(data.shape[0]): #data.shape[0] - кол-во строк(юзеров)к в массиве data...перебираем всех юзеров
    if (u != v): #чтобы не посчитать метрику сходства для самого себя
        uv_sum = 0 #сумма произведений u на v
        uu_sum = 0 #сумма квадратов u
        vv_sum = 0 #сумма квадратов v
        uv = data.iloc[[u, v]]#извлекаем строку пользователя u и v
        uv = uv.transpose()#транспонируем полученную матрицу (превращаем две строки в два столбца)
        for i,movie in uv.iterrows():#перебираем значения оценок попарно
            if ((movie[u] != without_evaluating) & (movie[v] != without_evaluating)):#если у обоих пользователей этот фильм оценен
                uv_sum += movie[u] * movie[v] #оценка пользователя u для определенного фильма и оценка нашего пользователя умножаются
                uu_sum += pow(movie[u], 2)#возводим в степень
                vv_sum += pow(movie[v], 2)#возводим в степень
        similarity_metric[u] = round(uv_sum / ((math.sqrt(uu_sum) * math.sqrt(vv_sum))), 3)#считаем мерику и округляем до 3х знаков после запятой
```

```python
similarity_metric
```

```python
sim4 = np.argpartition(similarity_metric, -k)[-k:]#функция расставляет в порядке возрастания и мы берем нужные нам последние 4 индекса пользователей
```

```python
sim4
```

```python
def means(user):
    l = []#пустой список
    for i in range(data.shape[1]):#data.shape[1] дает кол-во столбцов, т.е. фильмов
        if data.iloc[user][i] != without_evaluating:#пропускаем фильмы, к-е не оценены
            l.append(data.iloc[user][i])#добавляем значение каждого оцененного фильма
    arr = np.array(l)#превращаем список в массив arr
    return np.mean(arr)#вычисляем среднее арифметическое
```

```python
l = []
for i in range(data.shape[1]):
    if (data.iloc[v][i] == without_evaluating):
        l.append(i)
unrated_films = np.array(l)
unrated_films
```

```python
#rui
movie_ratings = {}#создаем словарь
for film in unrated_films:
    sumsimr = 0
    sumsim = 0
    for u in sim4:
        if (data.iloc[u, film] != without_evaluating):
            sumsimr += similarity_metric[u] * (data.iloc[u, film] - means(u))
            sumsim += similarity_metric[u]
    movie_ratings[film] = round(means(v) + sumsimr / sumsim, 3)
movie_ratings
```

```python
result_1 = {}
for i in movie_ratings.keys():
    result_1[f'Movie {i}'] = movie_ratings[i]
result_1
```

```python
context_day = context_day.drop(context_day.columns[0], axis = 1)
context_day
```

```python
#Удаляем столбец с юзерами
context_place = context_place.drop(context_place.columns[0], axis = 1)
context_place
```

```python
unrated_films
```

```python
#фильмы, просмотренные дома
home = ['h']
#фильмы, просмотренные на выходных
weekends = ['Sun','Sat']
film_rates = {}
for film in unrated_films:
    points_of_the_film = 0
    for u in sim4:
        if (data.iloc[u, film] != without_evaluating)&(context_place.iloc[u][film] in home)&(context_day.iloc[u][film] in weekends):
            points_of_the_film = points_of_the_film + movie_ratings[film]
    film_rates[film] = points_of_the_film
film_rates
```

```python
best_film = max(film_rates, key=film_rates.get)
movie_ratings[best_film]
```

```python
result_2 = {}
result_2[f'Movie {best_film}'] = movie_ratings[best_film]
result_2
```

```python
result = {
    'user': my_user,
    '1': result_1,
    '2': result_2
}

with open('result.json', 'w') as file:
    json.dump(result, file, indent = 4)
    
print(json.dumps(result, indent = 4))
```
